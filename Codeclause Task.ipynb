```python
#classifying movies based as rating
```

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
data["data_categories"]=data["Rating"].apply(rating)
```

```python
#Find unique values from Genre
```

```python
data["Genre"]
```

```python
list1=[]
for i in data["Genre"]:
    list1.append(i.split(","))
```

```python
list1
```

```python
oned=[]
for i in list1:
    for j in i:
        oned.append(j)
```

```python
oned
```

```python
uni=[]
for i in oned:
    if i not in uni:
        uni.append(i)
```
